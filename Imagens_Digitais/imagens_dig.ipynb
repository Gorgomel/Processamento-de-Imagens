```python
red[:, :, 1] = 0
red[:, :, 2] = 0
red[:, :, 0] = red[:, :, 0] %128
red_bgr = cv2.cvtColor(red, cv2.COLOR_RGB2BGR)
```

```python
blue[:, :, 0] = 0
blue[:, :, 1] = 0
blue[:, :, 2] = blue[:, :, 2] + 10
blue_bgr = cv2.cvtColor(blue, cv2.COLOR_BGR2RGB)
```
